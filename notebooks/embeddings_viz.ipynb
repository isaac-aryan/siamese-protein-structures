```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

print('Hi')

data = np.load("../data/processed/embeddings.npz")
X = data["embeddings"]
y = data["labels"]

X_2d = TSNE(n_components=2, perplexity=30).fit_transform(X)

plt.figure(figsize=(6, 6))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y, cmap="tab20", s=10)
plt.title("Protein Structure Embeddings")
plt.show()
```
